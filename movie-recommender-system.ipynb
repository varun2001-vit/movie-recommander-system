# Preprocessing

```python
movies.dropna(inplace=True)
```

```python
#we have to create 3 columns movie id , title ,tags which will come from keywords,cast,crew,and other remainging
```

```python
import ast
```

```python
#covertint string of list to list using ast and extracting the value for name in the dictionary
movies.iloc[0].genres
```

```python
# obj=[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):#converting string to list
        l.append(i['name'])#accessing each distictionary of list
    return l

    
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#same for keywords
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].cast
```

```python
movies['cast'].apply(convert)
```

```python
def convert_2(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 5:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert_2)
movies.head()
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            name=i['name']
            l.append(name)
    return l
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #splitting overview string to list
movies.sample(5)
```

# Removing spaces in the column values to avoid confusion between similar attribute values

```python
#now we need to remove spaces to avoid confusion with similar names in the dataset
#test
name="varun viswanathan"
name= name.replace(" ","")
name
```

## function to remove spaces

```python
def combine(obj):
    l=[]
    for i in obj:
        i=i.replace(" ","")
        l.append(i)
    return l

    
```

```python
movies['cast'].apply(combine)
```

```python
movies['cast'] = movies['cast'].apply(combine)
movies['crew'] = movies['crew'].apply(combine)
movies['genres'] = movies['genres'].apply(combine)
movies['keywords'] = movies['keywords'].apply(combine)
```

```python
movies.head()
```

# Creating the new column for tags which will have all the combined list values from all the columns 

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
new.head()
```

# There is a repeatation of words in the feature names therefore we are applying stemming 

```python
#applying stemming to the dataframe
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(object):
    l=[]
    for i in object.split():
        l.append(ps.stem(i))
    return " ".join(l)

```

```python
new.iloc[0].tags
```

```python
stem("in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver stephenlang michellerodriguez jamescameron")
```

```python
new['tags'].apply(stem)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
vector = cv.fit_transform(new['tags']).toarray()
vector
```

# calculate the cosine distance matrix between the vectors 

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

# Final Recommender function 

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

# Testing the function
